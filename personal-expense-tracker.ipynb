```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Create empty expense tracker
data = {
    "Date": [],
    "Category": [],
    "Amount": []
}

df = pd.DataFrame(data)

# Step 3: Function to add expenses
def add_expense(date, category, amount):
    global df
    new_entry = pd.DataFrame({"Date":[date], "Category":[category], "Amount":[amount]})
    df = pd.concat([df, new_entry], ignore_index=True)
    print(f"Added: {date}, {category}, {amount}₹")

# Step 4: Add sample expenses
add_expense("2026-02-01", "Food", 200)
add_expense("2026-02-02", "Transport", 150)
add_expense("2026-02-03", "Shopping", 500)
add_expense("2026-02-04", "Food", 250)
add_expense("2026-02-05", "Bills", 300)

# Step 5: View all expenses
print("\nAll Expenses:")
print(df)

# Step 6: Visualize total spending per category
category_sum = df.groupby("Category")["Amount"].sum().reset_index()

plt.figure(figsize=(6,4))
plt.bar(category_sum["Category"], category_sum["Amount"], color='skyblue')
plt.title("Expenses by Category")
plt.ylabel("Amount (₹)")
plt.show()

# Step 7: Visualize expenses over time
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(8,4))
plt.plot(df['Date'], df['Amount'], marker='o', linestyle='-', color='orange')
plt.title("Expenses Over Time")
plt.xlabel("Date")
plt.ylabel("Amount (₹)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
